```python
temp_df = df_renewable_elec_output.copy()
temp_df.head()
```

```python
countries=[]
for c in temp_df.columns:
    if c!='year':
        countries.append(c)        
#print (countries)  
```

```python
plt.figure(figsize=(14, 6))
for c1 in countries:
    temp_df_ = temp_df[[c1, 'year']]
    sns.lineplot(x=temp_df_["year"], y=temp_df_[c1], label=c1)
plt.xlabel("Year")
plt.ylabel("Efficiency") 
plt.title("Efficiency across the years per country")
```

```python
temp_df = df_renewable_elec_output[(1960 < df_renewable_elec_output.year) & (df_renewable_elec_output.year < 2016)]
temp_df = temp_df.drop(columns='year')
temp_df.describe()
```

```python
def plot_by_country(_df, title, nb_col):
    _df = _df.describe().iloc[nb_col, :]
    plt.figure(figsize=(22, 10))
    sns.barplot(x=_df.index, y=_df.values)
    plt.title(title)
```

```python
plot_by_country(temp_df, "75% efficiency by country", 6)
```

```python
def plot_corr(df_):
    corr = df_.corr()
    corr

    # Generate a mask for the upper triangle
    mask = np.zeros_like(corr, dtype=np.bool)
    mask[np.triu_indices_from(mask)] = True

    # Set up the matplotlib figure
    f, ax = plt.subplots(figsize=(14, 12))

    # Generate a custom diverging colormap
    #cmap = sns.diverging_palette(220, 10, as_cmap=True)

    # Draw the heatmap with the mask and correct aspect ratio
    sns.heatmap(corr, mask=mask, center=0, square=True, cmap='Spectral', linewidths=.5, cbar_kws={"shrink": .5}) #annot=True
```

```python
plot_corr(temp_df)
```
